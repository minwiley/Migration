```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
import json
import numpy as np
import pandas as pd

results_list = []

compound_list = []
positive_list = []
negative_list = []
neutral_list = []
tweets_data = []

def sentiment_analysis():
    for state in box.keys():
#         put all you state json file in a test file
        tweets_data_path = 'test/{}.json'
        tweets_file = open(tweets_data_path.format(state.rstrip()),'r')
        for line in tweets_file:
            try: 
                tweet = json.loads(line)
                tweets_data.append(tweet)
            except:
                continue
#         data = pd.read_csv('states_tweets/{}.json'.format(state.rstrip()))
            for i in range(len(tweets_data)):
                tweet_text = tweets_data[i]['text']
                results = analyzer.polarity_scores(tweet_text)
                compound = results["compound"]
                pos = results["pos"]
                neu = results["neu"]
                neg = results["neg"]
            
                compound_list.append(compound)
                positive_list.append(pos)
                negative_list.append(neg)
                neutral_list.append(neu)
        states_results = {"Username": state, "Compound Score": np.mean(compound_list),"Postive Score": np.mean(positive_list),"Neutral Score": np.mean(neutral_list),"Negative Score": np.mean(negative_list)}
        results_list.append(states_results)
        
        pd.DataFrame({'State':[state],'Compound':[np.mean(compound_list)],'Positive':[np.mean(positive_list)],'Negative':[np.mean(negative_list)]}).to_csv('{}.csv'.format(state.strip()))

        print(f"User: {state}")
        print(f"Compound: {np.mean(compound_list):.3f}")
        print(f"Positive: {np.mean(positive_list):.3f}")
        print(f"Neutral: {np.mean(neutral_list):.3f}")
        print(f"Negative: {np.mean(negative_list):.3f}")    
```
